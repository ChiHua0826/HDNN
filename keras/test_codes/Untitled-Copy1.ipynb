```python
#引用tensorflow函式庫
import tensorflow
#引用keras函式庫
import keras
#引用csv函式庫
import csv
#引用numpy函式庫
import numpy
#引用pyplot函式庫
import matplotlib.pyplot as plot

#讀取訓練資料
dataset = numpy.loadtxt("案例1.csv", delimiter=",")
X = dataset[:,0:1]
Y = dataset[:,1]
```

```python
#設定亂數種子
numpy.random.seed(0)

#設定神經網路結構
model = keras.models.Sequential()
model.add(keras.layers.Dense(1, input_dim = 1, activation = 'linear'))

#設定神經網路學習目標
model.compile(loss='mean_squared_error', optimizer = keras.optimizers.Nadam2(), metrics = ['mae'])
```

```python
#宣告繪製結果圖形之函式
def show_train_history(train_history, x1, x2):
    plot.plot(train_history.history[x1])
    plot.plot(train_history.history[x2])
    plot.title('Train History')
    plot.ylabel('train')
    plot.xlabel('Epoch')
    plot.legend([x1, x2], loc = 'upper right')
    plot.show()
```

```python
#訓練神經網路
train_history = model.fit(
    X, #輸入參數
    Y, #輸出參數
    epochs = 3000, #訓練回合數
    batch_size = 1) #逐筆修正權重

#顯示訓練過程
show_train_history(train_history, 'loss', 'mae')

#顯示訓練錯誤率
scores = model.evaluate(X, Y)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))

#顯示神經網路權重值
print(model.get_weights())
```

```python
#將測試資料代入模型進行預測,並取得預測結果
predictions = model.predict(X)
print(predictions)
```
