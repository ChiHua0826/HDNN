```python
import tensorflow
import keras
import csv
import numpy
import matplotlib.pyplot as plot

#loading training data
X = numpy.loadtxt("Datasets/Experiment1_x.csv", delimiter=",")
Y = numpy.loadtxt("Datasets/Experiment1_y.csv", delimiter=",")
```

```python
from keras.constraints import non_neg

#setting random seed
numpy.random.seed(0)

#constructing input layer
input_layer = keras.layers.Input(shape=(5, ), name='input_layer')

#constructing hidden layers
hidden_layer = keras.layers.Dense(1, activation = 'linear', name='hidden_layer', 
                                    use_bias=False, kernel_constraint=non_neg())(input_layer)

#constructing output layer
output_layer = keras.layers.Dense(1, activation = 'linear', name='output_layer')(hidden_layer)

#constructing the model of neural network
model = keras.models.Model(inputs=input_layer, outputs=output_layer)
model.summary()

#setting loss function and optimizer
model.compile(loss='mean_squared_error', optimizer = 'nadam', metrics = ['mae'])
```

```python
#training the model of neural network
train_history = model.fit(X, Y, epochs = 10000, batch_size = 7)
```

```python
#drawing the lines of losses and metrics
def show_train_history(train_history, x1, x2):
    plot.plot(train_history.history[x1])
    plot.plot(train_history.history[x2])
    plot.title('Train History')
    plot.ylabel('train')
    plot.xlabel('Epoch')
    plot.legend([x1, x2], loc = 'upper right')
    plot.show()
    
show_train_history(train_history, 'loss', 'mae')

#saing the weights of trained neural network
model.save_weights('Experiment1_model_weights.h5')

#showing the weights in the model of neural network
print(model.get_weights())
```
